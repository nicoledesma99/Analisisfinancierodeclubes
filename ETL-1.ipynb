```python
import pandas as pd
```

```python
# cargar el conjunto de datos
df = pd.read_csv("clubes.csv")

# filtrar los clubes que han ganado títulos de primera
df = df[df['Títulos_Primera'] > 0]

# seleccionar los 10 clubes con el peor ingreso total
df = df.nsmallest(10, 'Ingresos_Totales')

# guardar el nuevo conjunto de datos en un archivo csv
df.to_csv('clubes_historicos_con_porcentajes.csv', index=False)
```

```python
# Leer el archivo CSV con el dataset
df = pd.read_csv('clubes_historicos_con_porcentajes.csv')

# Eliminar columnas irrelevantes para el análisis
columnas_eliminar = ['ID_Club (FK)', 'Nombre_Coloquial', 'Afiliación_AFA']
df = df.drop(columnas_eliminar, axis=1)

# Rellenar valores nulos en columnas relevantes
columnas_relevantes = ['Títulos_Primera', 'Títulos_Internacionales']
df[columnas_relevantes] = df[columnas_relevantes].fillna(0)

# Convertir columnas de tipo objeto a tipo numérico
columnas_numericas = ['Títulos_Primera', 'Títulos_Internacionales']
df[columnas_numericas] = df[columnas_numericas].astype(int)

# Calcular el total de títulos de cada club
df['Total_Títulos'] = df['Títulos_Primera'] + df['Títulos_Internacionales']

# Eliminar filas con valores faltantes en columnas relevantes
df = df.dropna(subset=['Socios', 'Cuota_Social', 'Ingresos_Socios', 'Entradas_Vendidas'])

# Calcular nuevas columnas de interés
df['Ingresos_Total'] = df['Ingresos_Socios'] + df['Entradas_Vendidas'] + df['Porcentaje_Ingresos_TV'] + df['Porcentaje_Ingresos_Comerciales']
df['Porcentaje_Ingresos_Socios'] = (df['Ingresos_Socios'] / df['Ingresos_Total']) * 100
df['Porcentaje_Ingresos_Entradas'] = (df['Entradas_Vendidas'] / df['Ingresos_Total']) * 100



df.sample(10)

```

```python
# Guardar el resultado en un nuevo archivo CSV
df.to_csv('resultado_etl.csv', index=False)
```
